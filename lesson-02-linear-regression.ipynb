```python
df=pd.read_csv("/kaggle/input/salary-01/salary.csv")
df
```

```python
years_of_exp_val=df['YearsExperience'].values
years_of_exp_val
```

```python
salary=df['Salary'].values
salary
```

```python
from matplotlib import pyplot as plt
x=years_of_exp_val #x: Independent Variable (years of experience)
y=salary #y: dependent variable (salary)
```

```python
plt.scatter(x,y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.plot
```

# Create a column vector

```python
x=x.reshape(-1,1)
x, len(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=.8,random_state=100)
x_train, len(x_train)
```

```python
x_test,len(x_test)
```

```python
plt.scatter(x_train,y_train, color="red")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.plot
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
y_predict
```

```python
lr.score(x_test,y_test)*100
```
